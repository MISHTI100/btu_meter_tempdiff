```python
import pandas as pd


filenames = ['feb3_btu.xlsx', 'march3_btu.xlsx', 'march4_btu.xlsx', 'april3_btu.xlsx', 'april4_btu.xlsx', 'may3_btu.xlsx', 'may4_btu.xlsx', 'june3_btu.xlsx', 'june4_btu.xlsx', 'july3-aug_btu.xlsx', 'july4-aug_btu.xlsx']


df = pd.concat([pd.read_excel(file) for file in filenames], ignore_index=True)


df.head()
```

```python
df.drop(columns=['timestamp', 'location', 'name'], inplace=True, errors='ignore')
```

```python
df.head()
```

```python
df.tail()
```

```python
df = df[df[df.columns[0]].notnull()]
df = df[df[df.columns[0]] != 0]
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

df_filtered = df[df['temp_dif'] > 1]
plt.figure(figsize=(10, 6))
plt.plot(df_filtered.index, df_filtered['temp_dif'], label='Temperature Difference (C)', marker='o')
plt.xlabel('Index')
plt.ylabel('Temperature Difference (C)')
plt.title('BTU Meter - Temp Difference > 1C')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
